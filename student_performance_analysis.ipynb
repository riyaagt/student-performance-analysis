```python
import pandas as pd
df = pd.read_csv("Student_Performance.csv")
df.head()
```

```python
sns.scatterplot(
    data=df,
    x="study_hours",
    y="attendance_percentage"
)
plt.show()
```

```python
### Insight 1
Students who spend more hours studying generally tend to have higher attendance percentages.
```

```python
sns.boxplot(
    data=df,
    x="school_type",
    y="attendance_percentage"
)
plt.show()
```

```python
### Insight 2
Attendance percentage varies between public and private schools, suggesting that school type may influence student attendance.
```

```python
sns.boxplot(
    data=df,
    x="internet_access",
    y="study_hours"
)
plt.show()
```

```python
### Insight 3
Students with internet access show differences in study hours compared to those without internet access.
```

```python
## Conclusion

This analysis explored how study habits, attendance, school type, and internet access relate to student performance.
The findings suggest that consistent study habits and access to resources may positively influence attendance and learning outcomes.
```
